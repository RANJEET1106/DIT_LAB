# Data Encoders


## Label Encoder


```python
dflabel=copy.deepcopy(data)
```

```python
labelencoder_x=le()
dflabel['Sex']=labelencoder_x.fit_transform(dflabel['Sex'])
```

```python
dflabel.head(5)
```

```python
labelencoder_x=le()
dflabel['Embarked']=labelencoder_x.fit_transform(dflabel['Embarked'])
```

```python
dflabel.head(5)
```

## One Hot Encoder

```python
dfhotencoder=copy.deepcopy(data)
```

```python
result=ohe().fit_transform(dfhotencoder['Sex'].values.reshape(-1,1)).toarray()
dfhotencoder[['Female','Male']]=pd.DataFrame(result,index=dfhotencoder.index)
```

```python
dfhotencoder.drop('Sex',axis=1,inplace=True)
```

```python
dfhotencoder.head(5)
```

```python
result=ohe().fit_transform(dfhotencoder['Embarked'].values.reshape(-1,1)).toarray()
dfhotencoder[['C','Q','S']]=pd.DataFrame(result,index=dfhotencoder.index)
```

```python
dfhotencoder.drop('Embarked',axis=1,inplace=True)
```

```python
dfhotencoder.head(5)
```
